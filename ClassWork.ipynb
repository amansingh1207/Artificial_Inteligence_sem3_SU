```python
import random

NUM_TRIALS = 1000

def run_simulation():
    success_count = 0

    for _ in range(NUM_TRIALS):
        
        r1 = random.random()
        
        result1 = 1 if r1 < 0.5 else 0

        r2 = random.random()
        
        result2 = 1 if r2 < 0.4 else 0

        multiplication_result = result1 * result2

        if multiplication_result == 1:
            success_count += 1

    print("--- Random Probability Simulation ---")
    print(f"Number of Trials: {NUM_TRIALS}")
    print(f"Total times multiplication (1 * 1) was 1: {success_count}")
for i in range(1,11):
    print("Observation in round: ", i)
    run_simulation()
```
