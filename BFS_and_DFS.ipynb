```python
# --- Define the Graph ---
graph = {
    'u': ['v', 'x'],
    'v': ['y'],
    'w': ['y', 'z'],
    'x': ['v'],
    'y': ['x'],
    'z': ['z']  # self-loop
}

# --- Global variables for DFS ---
time = 0
color = {}
parent = {}
discovery = {}
finish = {}

# --- DFS Algorithm ---
def DFS(G):
    """
    Runs Depth-First Search on graph G.
    Initializes all vertices, then calls DFS-VISIT on unvisited ones.
    """
    global time, color, parent, discovery, finish

    # Initialize all vertices
    for u in G:
        color[u] = "WHITE"   # Unvisited
        parent[u] = None

    time = 0

    # Perform DFS on each unvisited node
    for u in G:
        if color[u] == "WHITE":
            DFS_VISIT(G, u)

def DFS_VISIT(G, u):
    """
    Visits a vertex u in DFS.
    Tracks discovery time, finish time, and parent relationships.
    """
    global time, color, parent, discovery, finish

    color[u] = "GRAY"      # Node is discovered but not finished
    time += 1
    discovery[u] = time    # Discovery time

    # Explore neighbors
    for v in G[u]:
        if color[v] == "WHITE":
            parent[v] = u
            DFS_VISIT(G, v)

    color[u] = "BLACK"     # Finished exploring u
    time += 1
    finish[u] = time       # Finish time


# --- Run DFS ---
DFS(graph)

# --- Print Results ---
print("Discovery Times:", discovery)
print("Finish Times:", finish)
print("Parents:", parent)
```

```python
import time

def read_graph_from_file(filename):
    graph = {}
    with open(filename, "r") as f:
        lines = f.read().strip().split("\n")

    # last line = source, destination
    src, dest = lines[-1].split()

    # build adjacency list
    for line in lines[:-1]:
        node, neighbors = line.split(":")
        node = node.strip()
        neighbors = neighbors.strip().split()
        graph[node] = neighbors

    return graph, src, dest


def bfs(graph, start, goal):
    visited = set()
    queue = [[start]]  # path ko store karne ke liye list of lists

    while queue:
        path = queue.pop(0)  # FIFO
        node = path[-1]

        if node == goal:
            return path

        if node not in visited:
            visited.add(node)
            for neighbor in graph.get(node, []):
                new_path = list(path)
                new_path.append(neighbor)
                queue.append(new_path)
    return None


def dfs(graph, start, goal):
    visited = set()
    stack = [[start]]  # path store karne ke liye

    while stack:
        path = stack.pop()  # LIFO
        node = path[-1]

        if node == goal:
            return path

        if node not in visited:
            visited.add(node)
            for neighbor in graph.get(node, []):
                new_path = list(path)
                new_path.append(neighbor)
                stack.append(new_path)
    return None


if __name__ == "__main__":
    graph, src, dest = read_graph_from_file("graph.txt")

    print("Graph:", graph)
    print(f"Source: {src}, Destination: {dest}\n")

    # BFS timing
    start_time = time.time()
    bfs_path = bfs(graph, src, dest)
    bfs_time = time.time() - start_time

    # DFS timing
    start_time = time.time()
    dfs_path = dfs(graph, src, dest)
    dfs_time = time.time() - start_time

    print("BFS Path:", bfs_path)
    print("DFS Path:", dfs_path)
    print("\n--- Time Comparison ---")
    print(f"BFS Time: {bfs_time} seconds")
    print(f"DFS Time: {dfs_time} seconds")
```
